```python
import os
import csv
import pandas

for post_id in post_ids:

    post_date = get_post_date(post_id, ACCESS_TOKEN).split("T")[0]
    print(post_date)

    # Get paths for the new folder and the master csv
    folder_path = f"DATA/{post_date}"
    csv_file_path = os.path.join(folder_path, "post-master.csv")

    os.makedirs(folder_path, exist_ok=True)

    # If first time date arrives
    if not os.path.exists(csv_file_path):
    # Step 3: Create and write to the CSV file
        data = {
            "POST_ID": [post_id],
            "CAPTION": [get_facebook_post_caption(post_id, ACCESS_TOKEN)]
        }
        df = pd.DataFrame(data)
        df.to_csv(csv_file_path, index = False)

        print(f"CSV file created at: {csv_file_path}")
    else:
        new_data = {
            "POST_ID": post_id,
            "CAPTION": get_facebook_post_caption(post_id, ACCESS_TOKEN)
        }
        df = pd.read_csv(csv_file_path)
        df.loc[len(df)] = new_data
        df.to_csv(csv_file_path, index = False)
        print(f"CSV file already exists at: {csv_file_path}")
    
```

```python
import os
folders = [name for name in os.listdir("DATA-large")]

for folder in folders:
    csv_dir = f"DATA-large/{folder}/post-master.csv"
    post_master_df = pd.read_csv(csv_dir)

    for post_id in post_master_df["POST_ID"]:
        print(post_id)     
        

        # Fetch all comments
        comments_data = get_all_comments(ACCESS_TOKEN, post_id)
        print("Got Comments")
        # print(comments_data)
        # exit()
        COMMENTS_AND_REPLIES = []
        COMMENT_AND_REPLIES_IDS = []
        
        # # Display comments and fetch replies
        for comment in comments_data:
            COMMENT_AND_REPLIES_IDS.append(comment['id'])
            COMMENTS_AND_REPLIES.append(comment['message'])

            replies = get_all_replies(ACCESS_TOKEN, comment['id'])
            
            if replies:
                for reply in replies:
                    COMMENT_AND_REPLIES_IDS.append(reply["id"])
                    COMMENTS_AND_REPLIES.append(reply['message'])
            else:
                None
        print("Got replies")

        REACTIONS = get_reaction_counts(post_id, ACCESS_TOKEN)

        data = {
            "COMMENT_ID" : COMMENT_AND_REPLIES_IDS,
            "COMMENT" : COMMENTS_AND_REPLIES,
            "REACTION_COUNT" : REACTIONS,
            "COMMENT_COUNT" : len(COMMENTS_AND_REPLIES)
        }
        post_df = pd.DataFrame(data)
        post_df.to_csv(f"DATA/{folder}/{post_id}.csv", index = False)

        print(f"Done at DATA/{folder}/{post_id}.csv comments = {len(COMMENTS_AND_REPLIES)}")




```
