```python
import pandas as pd

df = pd.read_csv("analyzed-data-new.csv")
df
```

```python
intents = ["Supportive or Appreciative", "Question or Inquiry", "Complaint or Dissatisfaction", 'Suggestion or Feedback', "General Conversation"]
intents
```

```python
import ast
irrelevant_count = 0
for index, row in df.iterrows():
    if row["RELEVANCE"] == "RELEVANT":
        post_id = row["POST_ID"]; caption = row["SUMMARY"]

        comment_ids = ast.literal_eval(row["COMMENT_IDS"])
        comments = ast.literal_eval(row["COMMENTS"])
        sentiments = ast.literal_eval(row["SENTIMENT"])
        intents = ast.literal_eval(row["INTENT"])

        print(f"POST {index} : ID - {post_id}\nCAPTION SUMMARY - {caption}\n")
        for comment_id, comment, sentiment, intent in zip(comment_ids, comments, sentiments, intents):
            print("ID: "+comment_id, "COMMENT: "+comment, "SENTIMENT: "+str(sentiment), "INTENT: "+intent, sep = "\n")
            print("\n")
        print("\n"*4)
    else:
        irrelevant_count += 1

```

```python
print(f"IRRELEVANT POSTS : {irrelevant_count}")
```
