```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("zero-shot-classification", model="MoritzLaurer/bge-m3-zeroshot-v2.0")
```

```python
import ast
INTENTS = ['lost', 'traffic', 'crime', 'political activities','missing']
for index, row in dataset.iterrows():
    if row["classification"] == "Eligible":
        # if count == 1:
        #     break
        # count += 1
        comments = ast.literal_eval(row["comments"])
        sentiment_labels = ast.literal_eval(row["sentiment"])

        intents = []
        # print(sentiment_labels)
        for idx, (comment, label) in enumerate(zip(comments, sentiment_labels)):
            if label == 1:
                row["intent"][idx] = "happy"
                continue
            output = pipe(comment, INTENTS)
            # intents.append()
            print(comment, output["labels"][0], sep = " = ")
            row["intent"][idx] = output["labels"][0]
        print("\n", intents)
        # dataset.iloc[index, -1] = intents
        
```
