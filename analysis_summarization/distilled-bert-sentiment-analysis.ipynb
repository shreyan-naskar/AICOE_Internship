```python
for _, row in dataset.iterrows():
    row["RELEVANCE"] = row["RELEVANCE"].rstrip()

dataset.to_csv("new_data_intent.csv", index = False)
```

```python
import numpy as np
import ast
dataset['SENTIMENT'] = dataset['COMMENTS'].apply(lambda x: ['' for _ in range(len(ast.literal_eval(x)))] if isinstance(x, str) else [])
dataset
```

```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis", model="distilbert/distilbert-base-uncased-finetuned-sst-2-english")
```

```python
import ast
count = 0
labels_map = {
    "POSITIVE": 1,
    "NEGATIVE": -1
}
# print("Hello")
for index, row in dataset.iterrows():
    # print("hello", row["RELEVANCE"], type(row["RELEVANCE"]))
    if row["RELEVANCE"] == "RELEVANT":
        # print("hello1")
        # if count == 1:
        #     break
        # count += 1
        texts = row["COMMENTS"]
        sentiments = []
        for i, text in enumerate(ast.literal_eval(texts)):
            outputs = classifier(text[:613])
            gen_label = outputs[0]["label"]
            # sentiments.append()
            row["SENTIMENT"][i] = labels_map[gen_label]
        print(index, "Done")

```
