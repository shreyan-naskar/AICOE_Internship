```python
import pandas as pd
dataset=pd.read_csv('new_data_intent.csv')
dataset
```

```python
dataset['SUMMARY'] = 'a'
dataset
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline
classifier = pipeline("summarization", model="Falconsai/text_summarization", max_length = 50)
```

```python
# allowed = list(range(ord("A"), ord("Z") + 1)) + list(range(ord("a"), ord("z") + 1)) + list(range(ord("0"), ord("9") + 1))
for index, row in dataset.iterrows():
    if row["RELEVANCE"] == "RELEVANT":
        # lines = row["text"].split("\n")
        # caption = ""
        # for line in lines:
        #     if line and ord(line[0]) in allowed :
        #         # print(line[0], line, "\n")
        #         caption += line

        summaryGen = classifier(row["CAPTIONS"][:512])
        print(index, "Done", type(summaryGen[0]["summary_text"]), summaryGen[0]["summary_text"])
        print("\n"*2)
        dataset.iloc[index, -1] = summaryGen[0]["summary_text"]
```
